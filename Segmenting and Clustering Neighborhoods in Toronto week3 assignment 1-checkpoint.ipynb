```python
import pandas as pd
import numpy as np
import matplotlib as pl
import matplotlib.pyplot as mp
%matplotlib inline 
```

```python
online_data=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
neighbor_toronto_data=online_data[0]
```

```python
dft=pd.DataFrame (neighbor_toronto_data)
```

```python
dft.head()
```

```python
data_toronto= dft[dft['Borough']!='Not assigned']
```

```python
data_toronto.head(20)
```

```python
data_toronto.groupby('Postcode')
```

```python
data_toronto.drop_duplicates()
```

```python
Toronot_coll=data_toronto.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ','.join(x))
```

```python
Toronot_coll
```

```python
Toronot_coll['Neighbourhood']=Toronot_coll['Neighbourhood'].replace('Not assigned', "Queen's Park")
```

```python
Toronot_coll.reset_index(inplace=True)
```
